```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
uploaded = files.upload()
```

```python
# Membuat variabel pada Dataset
ngambil_data = pd.read_csv('Data_curah_hujan.csv', sep=';')
```

```python

# Mengambil semua data
data_terambil = ngambil_data

# Plotting
plt.figure(figsize=(12, 6))

# Mengatur lebar bar
bar_width = 0.15

# Mengatur posisi bar untuk setiap periode Data curah hujan dengan urutan 1, 2, 3
Bulan = np.arange(len(data_terambil))
Data_Curah_hujan1 = Bulan
Data_Curah_hujan2 = Bulan + bar_width
Data_Curah_hujan3 = Bulan + 2 * bar_width

# Bar plot untuk Data curah hujan setiap periode
plt.bar(Data_Curah_hujan1, data_terambil['Curah Hujan 1981-1990'], color='blue', width=bar_width, label='1981-1990')
plt.bar(Data_Curah_hujan2, data_terambil['Curah Hujan 1991-2000'], color='red', width=bar_width, label='1991-2000')
plt.bar(Data_Curah_hujan3, data_terambil['Curah Hujan 2001-2010'], color='#008763', width=bar_width, label='2001-2010')

# Menghitung nilai maksimum dalam dataset
nilai_maksimal = data_terambil[['Curah Hujan 1981-1990', 'Curah Hujan 1991-2000', 'Curah Hujan 2001-2010']].max().max()

# Mengatur batas atas sumbu y menjadi 700
batas_atas_y = max(nilai_maksimal, 700)

# Menambahkan garis horizontal
for i in range(20, int(batas_atas_y), 20):
    warna = 'black' if i % 100 == 0 else 'gray'
    plt.axhline(y=i, color=warna, linestyle='-', zorder=0)

# Konfigurasi plot
plt.xlabel('Bulan')
plt.ylabel('Curah Hujan (mm)')
plt.ylim(0, batas_atas_y)
plt.xticks(Bulan + bar_width, data_terambil['Bulan'])

# Menampilkan legenda menyamping di tengah
plt.legend(bbox_to_anchor=(0.5, 1.15), loc='upper center', ncol=3)

# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
nampil_data = pd.read_csv('Celcius.csv')
print(nampil_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Membuat Variabel di masing-masing data dari dataset "Celcius.csv"
waktu = (nampil_data['waktu'])
data_asimilasi =  (nampil_data['asimilasi'])
data_observasi =  (nampil_data['observasi'])
data_non_asimilasi =  (nampil_data['non asimilasi'])

#Style
sns.set(style="darkgrid")

# Membuat grafik multi line plot
plt.figure(figsize=(10, 5))
plt.plot(waktu, data_asimilasi, marker='o', label='asimilasi', color="blue")
plt.plot(waktu, data_observasi, marker='o', label='observasi', color="orange")
plt.plot(waktu, data_non_asimilasi, marker='o', label='Non asimilasi', color="red")

# Menambahkan label pada sumbu x dan y
plt.xlabel('Tgl/Bln/Tahun & waktu')
plt.ylabel('Celcius')

#judul grafik
plt.title('Suhu Permukaan')

# Memutar label tanggal dan jam agar tidak saling berdekatan/bertumpangan
plt.xticks(rotation=80)
plt.yticks([20, 21, 22, 23, 24, 25, 26, 27])
plt.grid(False, axis='x')
plt.legend()
# Menampilkan grafik
plt.show()
```
